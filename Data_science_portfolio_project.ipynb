# 📑 Chapter 8: Handling Missing Values

## 📖 Introduction

Missing values are a common challenge in real-world datasets and must be addressed before performing statistical analysis or visualization. Incomplete data can reduce the reliability of analytical results, introduce bias, and affect the performance of predictive models.

There are several approaches for handling missing values, including removing incomplete observations, replacing missing values with statistical measures such as the mean or median, or using more advanced imputation techniques.

In this project, appropriate imputation techniques will be applied based on the characteristics of each variable to preserve as much information as possible while maintaining the integrity of the dataset.


## 🎯 Purpose

## 🔍 Section 8.1: Review Missing Values

Before replacing missing values, it is useful to identify which variables contain incomplete observations and determine the extent of missing data within each variable.

```python
# ==========================================================
# Display Missing Values
# ==========================================================

missing_data = df.isnull().sum().sort_values(ascending=False)

missing_data
```

## 📊 Result: Missing Values by Variable

The table below displays the total number of missing values identified in each variable.

```python
# ==========================================================
# Percentage of Missing Values
# ==========================================================

missing_percentage = (df.isnull().sum()/len(df))*100

missing_summary = pd.DataFrame({
    "Missing Values": df.isnull().sum(),
    "Percentage (%)": missing_percentage.round(2)
})

missing_summary.sort_values(by="Percentage (%)", ascending=False)
```

## 📊 Result: Missing Value Summary

The table below summarizes the number and percentage of missing values for each variable.


## 📖 Interpretation

Calculating missing value percentages provides a clearer understanding of data completeness. Variables with only a small percentage of missing values are generally suitable for statistical imputation, while variables with extensive missingness may require alternative treatment.

## 🔧 Section 8.2: Mean Imputation

For numerical variables containing relatively few missing values, replacing missing observations with the mean preserves the overall size of the dataset while minimizing the impact on the variable's distribution.

```python
# ==========================================================
# Replace Missing Values Using Mean
# ==========================================================

mean_bore = df["bore"].astype(float).mean()
df["bore"].fillna(mean_bore, inplace=True)

mean_stroke = df["stroke"].astype(float).mean()
df["stroke"].fillna(mean_stroke, inplace=True)

mean_horsepower = df["horsepower"].astype(float).mean()
df["horsepower"].fillna(mean_horsepower, inplace=True)

mean_peak_rpm = df["peak-rpm"].astype(float).mean()
df["peak-rpm"].fillna(mean_peak_rpm, inplace=True)

print("Numerical missing values successfully replaced.")
```

## 📊 Result: Numerical Missing Values Replaced

The missing numerical values have been replaced using the mean of their respective variables.

## 📖 Interpretation

Mean imputation preserves the number of observations available for analysis while maintaining the approximate central tendency of each variable.

## 🏷️ Section 8.3: Mode Imputation

Categorical variables cannot be imputed using numerical statistics such as the mean. Instead, the most frequently occurring category (mode) is commonly used.

```python
# ==========================================================
# Replace Missing Number of Doors
# ==========================================================

mode_doors = df["num-of-doors"].mode()[0]

df["num-of-doors"].fillna(mode_doors, inplace=True)

print("Categorical missing values successfully replaced.")
```

## 📊 Result: Categorical Missing Values Replaced

Missing values in the categorical variable have been replaced using the most frequently occurring category.

## 🗑️ Section 8.4: Remove Records with Missing Prices

The selling price is the primary variable of interest in this project. Observations without price information cannot contribute to the analysis and are therefore removed from the dataset.

```python
# ==========================================================
# Remove Missing Prices
# ==========================================================

df.dropna(subset=["price"], axis=0, inplace=True)

df.reset_index(drop=True, inplace=True)

print("Rows with missing prices successfully removed.")
```

## 📊 Result: Dataset Updated

Records containing missing selling prices have been removed, and the dataset index has been reset.


## 📖 Interpretation

Removing observations with missing values in the target variable ensures that all remaining records contain complete information required for price-related analyses.

## 📏 Section 9.3: Normalize Selected Variables

Normalization rescales variables to a common range, making them easier to compare. This technique is particularly useful when variables have very different numerical scales.

```python
# ==========================================================
# Normalize Length, Width and Height
# ==========================================================

df["length"] = df["length"] / df["length"].max()

df["width"] = df["width"] / df["width"].max()

df["height"] = df["height"] / df["height"].max()

df[["length", "width", "height"]].head()
```

## 📊 Result: Normalized Vehicle Dimensions

## 📖 Interpretation

Normalization removes differences in scale while preserving the relative relationships among observations. This makes variables easier to compare and is particularly useful for machine learning algorithms.

## 📊 Section 9.4: Create Horsepower Categories

Continuous numerical variables can be grouped into categories to simplify interpretation and facilitate comparison.

Horsepower will be classified into three categories:

- Low
- Medium
- High

```python
# ==========================================================
# Bin Horsepower
# ==========================================================

bins = np.linspace(df["horsepower"].min(),
                   df["horsepower"].max(),
                   4)

group_names = ["Low", "Medium", "High"]

df["horsepower-category"] = pd.cut(df["horsepower"],
                                   bins,
                                   labels=group_names,
                                   include_lowest=True)

df[["horsepower", "horsepower-category"]].head(10)
```

## 🔥 Section 10.11: Relationship Between Horsepower and Vehicle Price

While histograms describe the distribution of individual variables, they do not reveal relationships between variables. Scatter plots are widely used to examine the association between two numerical variables by displaying each observation as a point in a two-dimensional space.

In this section, horsepower is compared with vehicle price to determine whether more powerful vehicles generally command higher market prices.

```python
# ==========================================================
# Figure 10.8
# Horsepower vs Vehicle Price
# ==========================================================

plt.figure(figsize=(10,6))

plt.scatter(
    df["horsepower"],
    df["price"],
    alpha=0.7
)

# Best-fit regression line
m, b = np.polyfit(df["horsepower"], df["price"], 1)

plt.plot(
    df["horsepower"],
    m * df["horsepower"] + b,
    linewidth=2,
    label="Best-fit Line"
)

plt.title(
    "Figure 10.8: Relationship Between Horsepower and Vehicle Price",
    fontsize=14,
    fontweight="bold"
)

plt.xlabel("Horsepower (hp)", fontsize=12)

plt.ylabel("Vehicle Price (USD)", fontsize=12)

plt.legend()

plt.grid(alpha=0.3)

plt.tight_layout()

plt.show()
```

```python
# ==========================================================
# Pearson Correlation
# ==========================================================

correlation = df["horsepower"].corr(df["price"])

print(f"Pearson Correlation Coefficient: {correlation:.3f}")
```

## 📊 Result: Figure 10.8 – Horsepower vs Vehicle Price

The scatter plot displays the relationship between horsepower and vehicle price. Each point represents one vehicle, while the regression line summarizes the overall trend.

The Pearson correlation coefficient quantifies the strength and direction of the linear relationship between the two variables.

## 📖 Interpretation

The scatter plot should be examined for the overall direction, strength, and pattern of the relationship.

A positive upward trend indicates that vehicles with higher horsepower generally command higher prices. If the points cluster closely around the regression line, the relationship is relatively strong. Greater dispersion around the line suggests that additional variables also influence vehicle price.

The Pearson correlation coefficient provides numerical evidence of the relationship. Values close to **+1** indicate a strong positive linear association, values near **0** indicate little or no linear relationship, and values close to **−1** indicate a strong negative association.


## 💼 Business Interpretation

Horsepower is often regarded as a premium feature because it reflects vehicle performance. Manufacturers typically price higher-performance vehicles at a premium, resulting in a positive relationship between horsepower and selling price.

However, horsepower alone does not determine price. Brand reputation, engine technology, body style, luxury features, and market positioning also contribute significantly to vehicle valuation.


> ## 📝 Analyst's Note
>
> A strong correlation does **not** imply causation. Although horsepower and price may exhibit a strong association, this does not necessarily mean that horsepower alone causes higher prices. Other variables may influence both characteristics simultaneously.


> ## 📌 Best Practice

## 🏆 Section 10.19: Ranking Variables by Their Correlation with Vehicle Price

While the correlation matrix and heatmap provide a comprehensive overview of relationships among numerical variables, they can be difficult to interpret when many variables are present.

To simplify interpretation, this section ranks all numerical variables according to the absolute strength of their Pearson correlation with vehicle price. Ranking variables in this manner highlights the characteristics that exhibit the strongest linear associations with the target variable and provides valuable guidance for feature selection during predictive modeling.

```python
# ==========================================================
# Correlation Ranking with Vehicle Price
# ==========================================================

price_correlation = (
    df.select_dtypes(include="number")
      .corr()["price"]
      .drop("price")
      .sort_values(key=abs, ascending=False)
)

correlation_table = (
    price_correlation
    .reset_index()
)

correlation_table.columns = [
    "Variable",
    "Correlation with Price"
]

correlation_table.index = correlation_table.index + 1
correlation_table.index.name = "Rank"

display(correlation_table.round(3))
```

```python
# ==========================================================
# Figure 10.15
# Correlation Ranking
# ==========================================================

plt.figure(figsize=(10,7))

plt.barh(
    correlation_table["Variable"],
    correlation_table["Correlation with Price"]
)

plt.title(
    "Figure 10.15: Correlation of Numerical Variables with Vehicle Price",
    fontsize=14,
    fontweight="bold"
)

plt.xlabel("Pearson Correlation Coefficient", fontsize=12)

plt.ylabel("Variable", fontsize=12)

plt.gca().invert_yaxis()

plt.grid(axis="x", alpha=0.3)

plt.tight_layout()

plt.show()
```

## 📊 Result: Figure 10.15 – Correlation Ranking

The table and bar chart rank all numerical variables according to the strength of their linear relationship with vehicle price.

Variables appearing at the top of the ranking exhibit the strongest associations with vehicle price and therefore represent promising candidates for predictive modeling.

## 📖 Interpretation

The ranking provides a clear comparison of the explanatory power of each numerical variable with respect to vehicle price.

Variables with large positive correlation coefficients tend to increase as vehicle price increases, whereas variables with negative coefficients tend to decrease as price increases.

Analysts should pay particular attention to variables exhibiting the strongest correlations because they are likely to contribute significantly to regression models developed later in the project.


## 💼 Business Interpretation

The ranked correlations provide valuable insight into the primary drivers of vehicle pricing.

Characteristics exhibiting strong positive correlations may represent premium vehicle attributes that manufacturers use to differentiate products and justify higher prices. Conversely, variables with weak correlations may have relatively limited influence on pricing decisions.

These findings support evidence-based pricing strategies, product development, and market segmentation.


> ## 📝 Analyst's Note
>
> Variables with strong correlations should not automatically be included together in predictive models. Some predictors may themselves be highly correlated (multicollinearity), which can reduce model interpretability. Additional diagnostics will be performed during the modeling stage.


> ## 📌 Best Practice
>
> Ranking correlations with the target variable is a common feature selection technique during exploratory data analysis. It helps prioritize variables for further investigation while maintaining a transparent and systematic analytical workflow.
